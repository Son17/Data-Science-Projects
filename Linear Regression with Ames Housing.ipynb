```python
#separate data datatypes

objects = data.select_dtypes(include=['object'])
numerical = data.select_dtypes(exclude=['object'])
objects.head()
```

```python
#some of the numerical features are really categorical 

numerical = numerical.replace({"MS SubClass" : {20 : "SC20", 30 : "SC30", 40 : "SC40", 45 : "SC45", 
                                       50 : "SC50", 60 : "SC60", 70 : "SC70", 75 : "SC75", 
                                       80 : "SC80", 85 : "SC85", 90 : "SC90", 120 : "SC120", 
                                       150 : "SC150", 160 : "SC160", 180 : "SC180", 190 : "SC190"},
                               "Overall Qual": {10: "Very Excellent", 9: "Excellent", 8: "Very Good",
                                               7: "Good", 6: "Above Average", 5: "Average",
                                               4: "Below Average", 3: "Fair", 2: "Poor", 1: "Very Poor"},
                               "Overall Cond": {10: "Very Excellent", 9: "Excellent", 8: "Very Good",
                                               7: "Good", 6: "Above Average", 5: "Average",
                                               4: "Below Average", 3: "Fair", 2: "Poor", 1: "Very Poor"},
                               "Mo Sold" : {1 : "Jan", 2 : "Feb", 3 : "Mar", 4 : "Apr", 5 : "May", 6 : "Jun",
                                   7 : "Jul", 8 : "Aug", 9 : "Sep", 10 : "Oct", 11 : "Nov", 12 : "Dec"}
                              })
```

```python
#separate out additional object datatypes as defined above

add_objects = numerical[['MS SubClass', 'Overall Qual', 'Overall Cond', 'Mo Sold']]
numerical = numerical.drop(['MS SubClass', 'Overall Qual', 'Overall Cond', 'Mo Sold'], axis=1)
```

```python
#create new features with years

numerical['House_Age'] = numerical['Yr Sold'] - numerical['Year Built']
numerical['Modernised_Age'] = numerical['Yr Sold'] - numerical['Year Remod/Add']
#numerical['Garage_Age'] = numerical['Yr Sold'] - numerical['Garage Yr Blt']

#drop features with year in it
numerical = numerical.drop(['Yr Sold', 'Year Built', 'Yr Sold', 'Year Remod/Add'], axis=1)
```

```python
# Log transform of the skewed numerical features to reduce impact of outliers
# Inspired by Alexandru Papiu's script : https://www.kaggle.com/apapiu/house-prices-advanced-regression-techniques/regularized-linear-models
# As a general rule of thumb, a skewness with an absolute value > 0.5 is considered at least moderately skewed

from scipy.stats import skew

#log transform the target:
numerical["SalePrice"] = np.log1p(numerical["SalePrice"])

#log transform skewed numeric features:
numeric_feats = numerical.dtypes[numerical.dtypes != "object"].index

skewed_feats = numerical[numeric_feats].apply(lambda x: skew(x.dropna())) #compute skewness
skewed_feats = skewed_feats[skewed_feats > 0.75]
skewed_feats = skewed_feats.index

numerical[skewed_feats] = np.log1p(numerical[skewed_feats])
```

```python
#these are all categorical datatypes, therefore they will be transformed into dummy variables

object_dummies = pd.get_dummies(objects)
add_ojects_dummies = pd.get_dummies(add_objects)
```

```python
df_combi = pd.concat([numerical, object_dummies, add_ojects_dummies], axis=1)
df_combi.info()
```

```python
#double-check for missing data

df_combi[show_missing()].isnull().sum()
```

```python
#let's filter only select features with correlations above 0.25 or below -0.25

corr = df_combi.corr()

corr.index = df_combi.columns

filtered = corr['SalePrice'][(corr['SalePrice'] < -0.25) | (corr['SalePrice'] > 0.25)]
filtered_df = pd.DataFrame(filtered)
transposed = filtered_df.transpose()
cols = transposed.columns

print(str(filtered.shape[0]) + " features left with reasonnably high correlations to target.")
```

```python
#define X and y

X_raw = df_combi[cols]
index1 = data['SalePrice'].index
index2 = X_raw.index

y = data['SalePrice'][index2.isin(index1)]

X = X_raw.drop('SalePrice', axis=1)
X = X.replace(np.inf, 0)
X = X.fillna(X.median)
```

```python
#implement train/test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#build regression model

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X_train, y_train)

scorer = make_scorer(mean_squared_error)
mse_linear = cross_val_score(model, X_test, y_test, scoring=scorer, cv=5)

print("The MSE score for a linear regression model is {:.2f}.".format(mse_linear.mean()))
print("... this equates to a mean error of +/- {:.2f}.".format(np.sqrt(mse_linear.mean())))

scorer_r2 = make_scorer(r2_score)
mse_linear_r2_test = cross_val_score(model, X_test, y_test, scoring=scorer_r2, cv=5)
mse_linear_r2_train = cross_val_score(model, X_train, y_train, scoring=scorer_r2, cv=5)

print("The R2 score on the test data for a linear regression model is {:.2f}.".format(mse_linear_r2_test.mean()))
print("The R2 score on the training data for a linear regression model is {:.2f}.".format(mse_linear_r2_train.mean()))
```

```python
# build lasso regression model


from sklearn.linear_model import Lasso

lasso = Lasso(max_iter=100000)
lasso.fit(X_train, y_train)

mse_lasso = cross_val_score(lasso, X_test, y_test, scoring=scorer, cv=5)

print("The MSE score for a lasso regression model is {:.2f}.".format(mse_lasso.mean()))
print("... this equates to a mean error of +/- {:.2f}.".format(np.sqrt(mse_lasso.mean())))

mse_lasso_r2_test = cross_val_score(lasso, X_test, y_test, scoring=scorer_r2, cv=5)
mse_lasso_r2_train = cross_val_score(lasso, X_train, y_train, scoring=scorer_r2, cv=5)

print("The R2 score on the test data for a linear regression model is {:.2f}.".format(mse_lasso_r2_test.mean()))
print("The R2 score on the training data for a linear regression model is {:.2f}.".format(mse_lasso_r2_train.mean()))
```

```python
# build ridge regression model

from sklearn.linear_model import Ridge

ridge = Ridge(max_iter=10000)
ridge.fit(X_train, y_train)
              
mse_ridge = cross_val_score(ridge, X_test, y_test, scoring=scorer, cv=5)

print("The MSE score for a ridge regression model is {:.2f}.".format(mse_ridge.mean()))
print("... this equates to a mean error of +/- {:.2f}.".format(np.sqrt(mse_ridge.mean())))    

mse_ridge_r2_test = cross_val_score(ridge, X_test, y_test, scoring=scorer_r2, cv=5)
mse_ridge_r2_train = cross_val_score(ridge, X_train, y_train, scoring=scorer_r2, cv=5)

print("The R2 score on the test data for a linear regression model is {:.2f}.".format(mse_ridge_r2_test.mean()))
print("The R2 score on the training data for a linear regression model is {:.2f}.".format(mse_ridge_r2_train.mean()))
```

```python
# build elastic net regression model
    
from sklearn.linear_model import ElasticNet

en = ElasticNet(max_iter=10000)
en.fit(X_train, y_train)
              
mse_en = cross_val_score(en, X_test, y_test, scoring=scorer, cv=5)

print("The MSE score for a elastic net regression model is {:.2f}.".format(mse_en.mean()))
print("... this equates to a mean error of +/- {:.2f}.".format(np.sqrt(mse_en.mean())))     

mse_elastic_r2_test = cross_val_score(en, X_test, y_test, scoring=scorer_r2, cv=5)
mse_elastic_r2_train = cross_val_score(en, X_train, y_train, scoring=scorer_r2, cv=5)

print("The R2 score on the test data for a linear regression model is {:.2f}.".format(mse_elastic_r2_test.mean()))
print("The R2 score on the training data for a linear regression model is {:.2f}.".format(mse_elastic_r2_train.mean()))
```
