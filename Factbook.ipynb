```python
import pandas as pd
import sqlite3
import math
```

```python
conn = sqlite3.connect('factbook.db')

c = conn.cursor()
c.execute('SELECT * FROM facts limit 5;')

print(c.fetchall())
```

```python
area_land = c.execute('SELECT SUM(area_land) from facts WHERE area_land != "";').fetchall()
area_water = c.execute('SELECT SUM(area_water) from facts WHERE area_water != "";').fetchall()
land = pd.DataFrame(area_land)
water = pd.DataFrame(area_water)
print(land/water)
```
