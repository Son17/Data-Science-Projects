```python
data=pd.read_csv('funddata.csv')
```

```python
ax=data['avg'].plot(marker='o',color='r',linestyle='',label='average',figsize=(15,7),ylim=(0, 120),fontsize=15)
data['Benchmark'].plot(marker='o',color='k',linestyle='',label='benchmark',fontsize=15)
data.plot(x='Horizon',y='hi-lo',kind='bar',bottom=data['min'],ax=ax)
plt.ylabel('Performance %',fontsize=15)
plt.legend(loc=10,bbox_to_anchor=[0.5, -0.25],ncol=3,fontsize=15)
plt.title('UK Corporate Bond Universe',fontsize=15)
plt.show()
```
