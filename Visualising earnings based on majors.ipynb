```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.tools.plotting import scatter_matrix
import matplotlib
matplotlib.style.use('ggplot')
```

```python
recent_grads = pd.read_csv("C:\\Users\\sonja tilly.ABERDEEN\\data-master\\data-master\\college-majors\\recent-grads.csv")
recent_grads.head()
```

```python
recent_grads.describe()
```

```python
recent_grads = recent_grads.dropna()
```

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Median vs. Sample_size')
```

Students in popular subjects are not making more money than those in less popular ones.

```python
ax1 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax1.set_title('Unemployment_rate vs. Sample_size')
```

Sample size seems to range around 0.075 for the entire sample.

```python
ax2 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax2.set_title('Median vs. Full_time')
```

Most college majors seem to be earning between $0 and $50,000.

```python
ax3 = recent_grads.plot(x='Unemployment_rate', y='ShareWomen', kind='scatter')
ax3.set_title('ShareWomen vs. Unemployment_rate')
```

There is no strong link between the share of women in college majors and the unemployment rate.

```python
ax4 = recent_grads.plot(x='Median', y='Men', kind='scatter')
ax4.set_title('Median vs. Men')
```

```python
ax5 = recent_grads.plot(x='Median', y='Women', kind='scatter')
ax5.set_title('Median vs. Women')
```

The median salaries for women appear to to slightly lower than those for men, although there is no strong pattern.

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
color = ['cadetblue', 'magenta', 'gold', 'navy', 'purple']

fig = plt.figure(figsize=(5,20))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40, color=color[r])
```
