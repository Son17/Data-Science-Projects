```python
# Import libraries necessary for this project
import numpy as np
import pandas as pd

# Load the dataset
in_file = 'titanic_data.csv'
full_data = pd.read_csv('C:\\Users\\sonja tilly.ABERDEEN\\titanic_data.csv')
full_data1 = full_data.fillna(full_data.median())
# Print the first few entries of the RMS Titanic data
full_data1.head()
```

```python
# Store the 'Survived' feature in a new variable and remove it from the dataset
outcomes = full_data1['Survived']
data = full_data1.drop(['Survived', 'Ticket', 'Cabin', 'Name'], axis = 1)
data.head()
```

```python
# Convert categorical features

import pandas as pd
import numpy as np

df_with_dummies = pd.get_dummies(data[['Sex', 'Embarked']])  
data1 = pd.concat([data, df_with_dummies], axis = 1)
data2 = data1.drop(['Sex', 'Embarked'], axis = 1)
data2.head()
```

```python
# Create train test datasets
from sklearn.cross_validation import train_test_split

#train, test = train_test_split(data, train_size = 0.8)

X_train, X_test, y_train, y_test = train_test_split(data2, outcomes, test_size = 0.2, random_state = 25)
```

```python
# Create Gaussian naive bayes model

from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()
clf.fit(X_train, y_train)
result=clf.predict_proba(X_test).round(2)
clf.score(X_test, y_test)
```

```python
resultb=pd.DataFrame(result,columns=['Dead','Alive'],index=X_test['PassengerId'])
```
