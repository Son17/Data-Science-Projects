**Swarmplot**

This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
fig = plt.figure(figsize=(18,6))
sns.set_style("whitegrid")

sub1 = plt.subplot(1,4, 1)
sub1 = sns.swarmplot(y=df['1Y'], data=df)
plt.plot(df['1Y'][-1:].values, c='r', marker='*', markersize=20)

sub2 = plt.subplot(1,4, 2)
sub2 = sns.swarmplot(y=df['3Y'], data=df)
plt.plot(df['3Y'][-1:].values, c='r', marker='*', markersize=20)

sub3 = plt.subplot(1,4, 3)
sub3 = sns.swarmplot(y=df['5Y'], data=df)
plt.plot(df['5Y'][-1:].values, c='r', marker='*', markersize=20)

sub4 = plt.subplot(1,4, 4)
sub4 = sns.swarmplot(y=df['10Y'],data=df)
plt.plot(df['10Y'][-1:].values, c='r', marker='*', markersize=20)
```

**Boxplot**

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.Source: seaborn website

```python
sns.set_style("ticks")
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(111)
ax2 = fig.add_subplot(111)

df.boxplot(showmeans=True, whis = 100)

for i,d in enumerate(df):
    y = df[d]
    x = np.random.normal(i+1, 0.06, len(y))
    plt.plot(x, y, c = ["darkcyan","darkcyan","darkcyan", "darkcyan"][i], marker="*", linestyle='None',\
            label = '')

for i,d in enumerate(idx):
    y = idx[d]
    x = np.random.normal(i+1, 0.06, len(y))
    plt.plot(x, y, c='red', marker="*", markersize=10, linestyle='None', label='index')
    
plt.ylabel('Performance %',fontsize=12)
plt.title('Global Corporate Bond Universe',fontsize=12)

import matplotlib.lines as mlines
import matplotlib.pyplot as plt

red_line = mlines.Line2D([], [], color='red', marker='*',
                          markersize=20, label='index')
plt.legend(handles=[red_line])

plt.show()
```
